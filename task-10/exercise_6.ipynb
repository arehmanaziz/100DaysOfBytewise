# 6. Feature Engineering in the Heart Disease Dataset

Importing the libraries

```python
import pandas as pd
```

Loading the dataset

```python
df = pd.read_csv('./datasets/heart.csv')
```

```python
df.head()
```

Age Groups

```python
bins = [29, 39, 49, 59, 69, 79]
labels = ['30-39', '40-49', '50-59', '60-69', '70-79']
df['Age Group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

Cholesterol Levels

```python
chol_bins = [0, 200, 240, float('inf')]
chol_labels = ['Normal', 'Borderline', 'High']
df['Cholesterol Level'] = pd.cut(df['chol'], bins=chol_bins, labels=chol_labels)
```

Blood Pressure Categories

```python
bp_bins = [0, 120, 140, 160, float('inf')]
bp_labels = ['Normal', 'Elevated', 'Hypertension Stage 1', 'Hypertension Stage 2']
df['Blood Pressure Category'] = pd.cut(df['trestbps'], bins=bp_bins, labels=bp_labels)
```

Excersie Angina

```python
df['Exercise Angina'] = df['exang'].map({0: 'No', 1: 'Yes'})
```

Interaction Features

```python
df['Age_Chol_Interaction'] = df['age'] * df['chol']
df['BP_Exang_Interaction'] = df['trestbps'] * df['exang']
```
